```python
import pandas as pd
import subprocess
from datetime import datetime
```

```python
df = pd.read_csv('valid_runs.csv')
```

```python
df
```

```python
script = ['python', '/Users/florianmende/Documents/Development/thesis/implementation/__main__.py',
          '--compilation_repair_rounds 3 --execution_repair_rounds 2']

# get current datetime as id
id = datetime.now().strftime("%Y%m%d-%H%M%S")

for i in range(len(df)):
    for method in list(filter(None, df.iloc[i]["missing_methods"].strip("[]").split(" "))):
        run_script = script.copy()
        run_script.append(f"--run_id {df.iloc[i]['file'].strip('.csv')}")
        run_script.append(f"--method_range {method}:{method}")

        run_command = " ".join(run_script)

        # write to txt file
        with open(f"run_scripts/get_missing_methods_{id}.txt", "a") as f:
            f.write(run_command)
            f.write("\n")
```
